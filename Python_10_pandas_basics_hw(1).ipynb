**Задание 2**

```python
data = pd.read_csv('power.csv')
data.head()
#data [ data ['country'] == 'Lithuania']['quantity'].value_counts().head(1)
```

```python
df = data[(data['country'] == 'Lithuania') | (data['country'] == 'Estonia') | (data['country'] == 'Latvia')]
df['country'].unique()
```

```python
top =df[(df['category']==4) | (df['category']==12) | (df['category'] == 21) & (df['year'] >= 2005) & (df['year']<=2010)]
print("Cуммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года = ", top['quantity'].sum())
```
